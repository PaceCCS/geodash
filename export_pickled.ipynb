```python
import pickle
import torch
```

```python
# pickled_path = "../michal/ML_pytroch/VIS_25_09_2024/model_regression_PH_RS"
pickled_path = "./model_regression_PH_RO"
out_filename = "PH_RO.onnx"

# Load the pickled model
xtm, xts, yst, model, ro_min = pickle.load(open(pickled_path, 'rb'))

# Get input size from the normalization tensors
input_size = len(xtm)  # or len(xts)

# Set model to evaluation mode
model.eval()

# Export with dynamic batch dimension
example_input = torch.rand(1, input_size)
torch.onnx.export(
    model, 
    example_input, 
    out_filename,
    input_names=['input'],
    output_names=['output'],
    dynamic_axes={
        'input': {0: 'batch_size'},
        'output': {0: 'batch_size'}
    }
)

print(f"Model exported to {out_filename} with input size {input_size} and dynamic batch dimension")
```
